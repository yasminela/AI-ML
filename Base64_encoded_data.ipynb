```python
def vigenere_decrypt(ciphertext, key):
    decrypted = []
    key = key.lower()
    key_index = 0
    key_length = len(key)

    for char in ciphertext:
        if char.isalpha():
            shift = ord(key[key_index]) - ord('a')
            if char.islower():
                decrypted.append(chr((ord(char) - shift - ord('a')) % 26 + ord('a')))
            else:
                decrypted.append(chr((ord(char) - shift - ord('A')) % 26 + ord('A')))
            key_index = (key_index + 1) % key_length
        else:
            decrypted.append(char)

    return ''.join(decrypted)

ciphertext = "Zld hvv rse idgpie tptnh d pfhqwp eo uki biokgee Chgfyfw rsez kifvq lfp aepms tnfcw wbv ss eagrses oiai"
key = "defensylab"
print(vigenere_decrypt(ciphertext, key))

```

```python
import base64

# Base64 encoded data
encoded_data = "CTI2PC4TdwkyMzwuCig1LgkyMzwuaTkJMyN8Lg5raMLt"
decoded_data = base64.b64decode(encoded_data)

# Function to XOR decrypt
def xor_decrypt(data, key):
    return ''.join(chr(b ^ key) for b in data)

# Try all possible keys
for key in range(256):
    decrypted = xor_decrypt(decoded_data, key)
    print(f"Key {key}: {decrypted}")
```
